```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
df = pd.read_csv("train (1).csv")
df.head()
```

```python
df.shape
df = df.dropna()
```

```python
from sklearn import preprocessing
home_encoder = preprocessing.LabelEncoder()
home_encoder.fit(['Earth','Europa','Mars'])
df.iloc[:,1] = home_encoder.transform(df.iloc[:,1])
cryo_encoder = preprocessing.LabelEncoder()
cryo_encoder.fit(['True','False'])
df.iloc[:,2] = cryo_encoder.transform(df.iloc[:,2])
vip_encoder = preprocessing.LabelEncoder()
vip_encoder.fit(['True','False'])
df.iloc[:,6] = vip_encoder.transform(df.iloc[:,6])
df.head()
```

```python
transport_encoder = preprocessing.LabelEncoder()
transport_encoder.fit(['True','False'])
df.iloc[:,13] = transport_encoder.transform(df.iloc[:,13])
df.head()
```

```python
x = df[['ShoppingMall','RoomService','Spa','VRDeck','FoodCourt']].values
y = df[['Transported']].values
```

```python
from sklearn.ensemble import ExtraTreesClassifier
erf = ExtraTreesClassifier(n_estimators = 200)
erf = erf.fit(x,y)
yhat = erf.predict(x)
```

```python
from sklearn.metrics import f1_score
f1_score(y, yhat, average='weighted') 
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(learning_rate=0.1,max_features = 1,subsample = 0.5,n_estimators = 200)
gbc = gbc.fit(x,y)
yhat =gbc.predict(x)
f1_score(y, yhat, average='weighted')
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
abc = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(),learning_rate =0.1,n_estimators = 200)
abc = abc.fit(x,y) 
yhat =abc.predict(x)
f1_score(y, yhat, average='weighted')
```

```python
tdf = pd.read_csv('test (1).csv')
tdf.head()
```

```python
tdf['VRDeck'].isnull().values.any()
tdf['HomePlanet'] = tdf['HomePlanet'].fillna('Earth')
tdf['CryoSleep'] = tdf['CryoSleep'].fillna('False')
tdf['Age'] = tdf['Age'].replace(np.NaN,tdf['Age'].mean())
tdf['RoomService'] = tdf['RoomService'].replace(np.NaN,tdf['RoomService'].mean())
tdf['FoodCourt'] = tdf['FoodCourt'].replace(np.NaN,tdf['FoodCourt'].mean())
tdf['Spa'] = tdf['Spa'].replace(np.NaN,tdf['Spa'].mean())
tdf['VRDeck'] = tdf['VRDeck'].replace(np.NaN,tdf['VRDeck'].mean())
tdf['VIP'] = tdf['VIP'].fillna('False')
tdf['ShoppingMall'] = tdf['ShoppingMall'].replace(np.NaN,tdf['ShoppingMall'].mean())
```

```python
from sklearn import preprocessing
home_encoder = preprocessing.LabelEncoder()
home_encoder.fit(['Earth','Europa','Mars'])
tdf.iloc[:,1] = home_encoder.transform(tdf.iloc[:,1])
cryo_encoder = preprocessing.LabelEncoder()
cryo_encoder.fit(['True','False'])
tdf.iloc[:,2] = cryo_encoder.transform(tdf.iloc[:,2])
vip_encoder = preprocessing.LabelEncoder()
vip_encoder.fit(['True','False'])
tdf.iloc[:,6] = vip_encoder.transform(tdf.iloc[:,6])
tdf.head()
```

```python
x_t = tdf[['ShoppingMall','RoomService','Spa','VRDeck','FoodCourt']].values
l = list(np.concatenate(tdf[['PassengerId']].values).flat)
```

```python
y_t = erf.predict(x_t)
ERF = pd.DataFrame(y_t,columns = ['Transported'])
ERF.insert(0,"PassengerId",l, True)
ERF['Transported'] = ERF['Transported'].replace(0,'False')
ERF['Transported'] = ERF['Transported'].replace(1,'True')
ERF.head()
```

```python
ERF.to_csv('/content/prediction.csv',index = False)
```
